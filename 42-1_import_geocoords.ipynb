```python
import numpy as np
import pandas as pd
import csv
```

## Create a file with business name, longitudinal, and latitude

```python
df = pd.read_csv('data/ready4model_v2.csv')
```

```python
with open('data/geo_coords_sf.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        address = row[0]
        longitude = row[1]
        latitude = row[2]
        n = len(df)
        for idx in range(n):
            df_address = df.loc[idx,'business_address']
            if df_address == address[:-6]:  # To remove ' SF CA' at the end of addresses
                df.loc[idx,'business_longitude'] = longitude
                df.loc[idx,'business_latitude'] = latitude
```
