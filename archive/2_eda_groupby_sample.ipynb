### Count the number of violations during the period and fill the idf table.

#### periods
1. p1_3: 1 - 3 months before  
2. p4_6: 4 - 6 months before  
3. p7_9: 7 - 9 months before  
4. p10_12: 10 - 12 months before  
5. p13_18: 13 - 18 months before  
6. p19_24: 19 - 24 months before  
7. p25_36: 25 - 36 months before  

```python
for row in df_grouped.iterrows():
    bid = row[0][0]
    date = row[0][1]
    if 20180101 <= date <= 20180331:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p1_3'] = num_viols
    elif 20171001 <= date <= 20171231:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p4_6'] = num_viols
    elif 20170701 <= date <= 20170930:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, '7_9'] = num_viols
    elif 20170401 <= date <= 20170630:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p10_12'] = num_viols
    elif 20151001 <= date <= 20150331:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p13_18'] = num_viols
    elif 20150401 <= date <= 20150930:
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p19_24'] = num_viols
    else: 
        num_viols = len(row[1]['short_violation_id'])
        idx = idf.business_id[idf.business_id ==bid].index.tolist()
        idf.at[idx, 'p25_36'] = num_viols
```
